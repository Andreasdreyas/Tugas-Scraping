```python
import requests
from bs4 import BeautifulSoup

root = 'https://jebreeetmedia.com/category/sepakbola-internasional/'
judul = []
liga = []
tanggal = []
paragraf = []

for page in range(1, 11):
    url = requests.get(root + 'page/' + str(page))
    soup = BeautifulSoup(url.text, 'html.parser')

    for content in soup.find_all("div", {"class": "content"}):
        title = content.find("h2", {"class": "is-title post-title"})
        paragraph = content.find("p")
        league = content.find("span", {"class": "meta-item post-cat"})
        date = content.find("span", {"class":"meta-item date"})

        if title and league and date and paragraph:
            judul.append(title.text.strip())
            paragraf.append(paragraph.text.strip())
            liga.append(league.text.strip())
            tanggal.append(date.text.strip())

import pandas as pd
df = pd.DataFrame(
    {'judul':judul,
     'paragraf' : paragraf,
     'liga':liga,
     'tanggal' : tanggal
     })
df
```
